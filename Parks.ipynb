```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv("Parks.csv", index_col="id")
```

```python
diffTypes = df['type'].unique()
y = df['type'].value_counts().sort_values()
y.plot(kind='bar', rot=0, figsize=(15,35))
print(20)
```

```python
#Creating a mask for each sport to count how many parks are in each neighborhood
#Tennis Mask
tennis_mask = df['type'] == 'Tennis'
tennis_data = df[tennis_mask]
tennis_counts = tennis_data['neighborhood'].value_counts()
#Hockey Mask
hockey_mask = df['type'] == 'Hockey'
Hockey_data = df[hockey_mask]
hockey_counts = Hockey_data['neighborhood'].value_counts()
#Basketball Half Court Mask
bball_half_mask = df['type'] == 'Basketball (Half)'
bball_half_data = df[bball_half_mask]
bball_half_counts = bball_half_data['neighborhood'].value_counts()
#Basketball Full Court Mask
bball_full_mask = df['type'] == 'Basketball (Full)'
bball_full_data = df[bball_full_mask]
bball_full_counts = bball_full_data['neighborhood'].value_counts()
#Pickleball Mask
pickleball_mask = df['type'] == 'Pickleball'
pickleball_data = df[pickleball_mask]
pickleball_counts = pickleball_data['neighborhood'].value_counts()
#Bocce Mask
bocce_mask = df['type'] == 'Bocce'
bocce_data = df[bocce_mask]
bocce_counts = bocce_data['neighborhood'].value_counts()
#Street Hockey Mask
street_hockey_mask = df['type'] == 'Street Hockey'
street_hockey_data = df[street_hockey_mask]
street_hockey_counts = street_hockey_data['neighborhood'].value_counts()
#General Sports Mask
GS_mask = df['type'] == 'General Sports'
GS_data = df[GS_mask]
GS_counts = GS_data['neighborhood'].value_counts()
#Volleyball Mask
vball_mask = df['type'] == 'Volleyball'
vball_data = df[vball_mask]
vball_counts = vball_data['neighborhood'].value_counts()
#Dek Hockey Mask
dek_hockey_mask = df['type'] == 'Dek Hockey'
dek_hockey_data = df[dek_hockey_mask]
dek_hockey_counts = dek_hockey_data['neighborhood'].value_counts()
#Horse Shoe Pitt Mask
hrseSh_mask = df['type'] == 'Horseshoe Pit'
hrseSh_data = df[hrseSh_mask]
hrseSh_counts = hrseSh_data['neighborhood'].value_counts()
#Lawn Bowling Mask
lawnBowling_mask = df['type'] == 'Lawn Bowling'
lawnBowling_data = df[lawnBowling_mask]
lawnBowling_counts = lawnBowling_data['neighborhood'].value_counts()
```

```python
df.groupby('type')['neighborhood'].value_counts().sort_values(ascending = True).plot(kind='barh', rot=0, figsize=(15,35))
```

```python
#Using concat combines above sorted data into neighborhoods and counts of parks
neighborhood_counts = pd.concat([tennis_counts, hockey_counts, bball_half_counts, bball_full_counts, pickleball_counts, bocce_counts, street_hockey_counts, GS_counts, vball_counts, dek_hockey_counts, hrseSh_counts, lawnBowling_counts])
#Graphs combined data
neighborhood_counts.sort_values(ascending = True).plot(kind='barh', rot=0, figsize=(15,35))
```

```python
# Renames and reads Playgrounds.csv
playgrounds = pd.read_csv("Playgrounds.csv", index_col="id")
# Plots a graph where the value_counts() gets how many parks in the neighborhood and renames the graph as ax for axis labeling
ax = playgrounds['neighborhood'].value_counts().plot(kind='barh', rot=0, figsize=(15,35))
ax.set_xlabel("Number of Playgrounds")
ax.set_ylabel("Name of Neighborhood")
plt.title("Playgrounds in Neighborhoods in Pittsburgh")
```

```python
total_Parks = pd.concat([neighborhood_counts, playgrounds['neighborhood'].value_counts()])
total_Parks.sort_values().tail(5).plot(kind='bar', rot=90, figsize=(5,15))
```
